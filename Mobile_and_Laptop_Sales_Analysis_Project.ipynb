```python
# Data Visualisation
# barchart Brandwise sales
brand_qty=df.groupby("Brand")["Quantity Sold"].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(brand_qty["Brand"], brand_qty["Quantity Sold"])
plt.xlabel("Brand")
plt.ylabel("Total Quantity Sold")
plt.title("Brand vs Total Quantity Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Filter only Mobile Phones
phones = df[df['product'].str.lower() == "mobile phone"]

# Calculate brand-wise total quantity sold
brand_sales = phones.groupby('brand')['price'].sum().reset_index()

# Sort for better visualization
brand_sales = brand_sales.sort_values(by='price', ascending=False)

# Bar chart
plt.figure(figsize=(10,6))
plt.bar(brand_sales['brand'], brand_sales['price'],color='#FE019A')

plt.xlabel("Brand")
plt.ylabel("Total prices ")
plt.title("Brand-wise Sales (Mobile Phones)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# region wise revenue
df["Revenue"] = df["Price"] * df["Quantity Sold"]

region_rev = df.groupby("Region")["Revenue"].sum()

plt.figure(figsize=(8,8))
plt.pie(region_rev.values,
        labels=region_rev.index,
        autopct='%1.1f%%',
        startangle=140)

plt.title("Region-wise Revenue Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Filter only Mobile Phones
phones = df[df['product'].str.lower() == "laptop"]

# Calculate brand-wise total quantity sold
brand_sales = phones.groupby('brand')['price'].sum().reset_index()

# Sort for better visualization
brand_sales = brand_sales.sort_values(by='price', ascending=False)

# Bar chart
plt.figure(figsize=(10,6))
plt.bar(brand_sales['brand'], brand_sales['price'],color='g')

plt.xlabel("Brand")
plt.ylabel("Total prices ")
plt.title("Brand-wise Sales (Laptop)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Yearwise sales
# Convert date
df["Dispatch Date"] = pd.to_datetime(df["Dispatch Date"])

# Extract year
df["Year"] = df["Dispatch Date"].dt.year

# Group by year
yearly_sales = df.groupby("Year")["Quantity Sold"].sum().sort_index()

# Vertical bar chart
plt.figure(figsize=(10,6))
plt.bar(yearly_sales.index, yearly_sales.values)   # Vertical bars
plt.xlabel("Year")
plt.ylabel("Total Quantity Sold")
plt.title("Year-wise Sales ( Bar Chart)")
plt.xticks(yearly_sales.index)
plt.tight_layout()
plt.show()
```

```python
# monthly sales
df["Dispatch Date"] = pd.to_datetime(df["Dispatch Date"])

# Extract Year-Month
df["Year_Month"] = df["Dispatch Date"].dt.to_period("M").astype(str)

# Group by Month
monthly_sales = df.groupby("Year_Month")["Quantity Sold"].sum().sort_index()

# Plot line chart
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o',color='orange')
plt.xlabel("Month")
plt.ylabel("Quantity Sold")
plt.title("Monthly Sales Trend (Quantity Sold)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# histogram
plt.hist(df['Processor Specification'],bins=10,edgecolor='black',color='green')
plt.xlabel('Processor')
plt.ylabel('Frequency')
plt.title('Processor Frequency')
plt.xticks(rotation=90)
plt.show()
```

## **RAMsize Wise sales**




```python
# RAM Size Wise sales
# Group by Customer Location
ram_sales = df.groupby("RAM")["Quantity Sold"].sum()

# Pie chart
plt.figure(figsize=(9,9))
plt.pie(ram_sales.values,labels=ram_sales.index,autopct="%1.1f%%",startangle=140)

plt.title("RAM_Size-wise Sales (Quantity Sold)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Heatmap for RAM,ROM,SSD
# ----- Clean RAM -----
df["RAM_clean"] = (df["RAM"].str.replace("GB", "", case=False).str.strip())
df["RAM_clean"] = pd.to_numeric(df["RAM_clean"], errors="coerce")

# ----- Clean ROM -----
df["ROM_clean"] = (df["ROM"].str.replace("GB", "", case=False).str.replace("TB", "000", case=False) )  # Convert 1TB  1000GB approx.str.strip())
df["ROM_clean"] = pd.to_numeric(df["ROM_clean"], errors="coerce")

# ----- Clean SSD -----
df["SSD_clean"] = (df["SSD"].str.replace("GB", "", case=False).str.replace("TB", "000", case=False).str.strip())
df["SSD_clean"] = pd.to_numeric(df["SSD_clean"], errors="coerce")

# Correlation matrix
corr = df[["RAM_clean", "ROM_clean", "SSD_clean"]].corr()

# ----- Heatmap -----
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap: RAM, ROM, SSD")
plt.show()
```

```python
# Top 10 Mobile phones and laptops
# Clean column names
df.columns = df.columns.str.strip()

# Convert Quantity Sold to numeric
df['Quantity Sold'] = pd.to_numeric(df['Quantity Sold'], errors='coerce')

# Top 10 Mobile Phone sales
top_mobile = (
    df[df['Product'] == 'Mobile Phone']
    .groupby('Brand')['Quantity Sold']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Top 10 Laptop sales
top_laptop = (
    df[df['Product'] == 'Laptop']
    .groupby('Brand')['Quantity Sold']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Create figure with two bar charts
plt.figure(figsize=(14,5))

# Mobile Phone bar chart
plt.subplot(1, 2, 1)
plt.bar(top_mobile.index, top_mobile.values,color='purple')
plt.title("Top 10 Mobile Phone Sales")
plt.xlabel("Brand")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)

# Laptop bar chart
plt.subplot(1, 2, 2)
plt.bar(top_laptop.index, top_laptop.values,color='green')
plt.title("Top 10 Laptop Sales")
plt.xlabel("Brand")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
